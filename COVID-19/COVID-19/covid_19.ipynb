# Introduction
The 'Bed Surge' data set is a data set that provides evidence for one of two things: </p>
1. The spike of COVID-19 is dramatically decreasing, and fewer people are needing to be admitted to hospitals
2. COVID-19 is clearly not exterminated, and select hospitals can't provide admissions for COVID patients

## Examination of the 'Bed Surge' Data Set

```python
covid_data = pd.read_csv('bed_surge.csv')
covid_data.head()
```

## Statistical Summary: Full Data

```python
covid_data.describe()
```

## Visualization of the Statistics

```python
sns.boxplot(data=covid_data)
plt.title("Bed Surge: Full Data")
plt.xlabel("Beds Statistics")
plt.ylabel("Beds Available")
```

## Interpretation of Visualization
Upon thorough review of the visualization of the statistics, you may see that there is a surplus of beds
compared to the amount ready to accept patients, and patients currently in bed. Noticeably, the beds in warm
status also indicates that many beds satisfy the beds in warm.
